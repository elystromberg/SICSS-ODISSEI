```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import datetime


#os.getcwd()

# Load data 
df = pd.read_csv("data_subsample.csv")

# Convert "date" column to datetime objects
df['cf19l393'] = pd.to_datetime(df['cf19l393'])

# Get unique years
unique_years = df['cf19l393'].dt.year.unique()

# Check if all dates are from the same year
if len(unique_years) == 1:
    print("All dates are from the same year:", unique_years[0])

    # Extract dates as strings in "mm-dd" format
    df['formatted_date'] = df['cf19l393'].dt.strftime('%m-%d')

    # Count the frequency of each date
    date_counts = df['formatted_date'].value_counts().sort_index()

    # Create a histogram
    date_counts.plot(kind='bar')

    # Customize the plot
    plt.xlabel('Date')
    plt.ylabel('Frequency')
    plt.title('Response Date Histogram')

    # Rotate the x-axis labels for better visibility
    plt.xticks(rotation=45)

    # Show only every second x-axis tick label
    plt.gca().xaxis.set_major_locator(plt.MultipleLocator(2))

    # Show the plot
    plt.show()
    
else:
    print("Dates are from different years")
```
